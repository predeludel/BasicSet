
### 1 - BlanklineTokenizer
Разбивает текст, рассматривая в качестве разделителя любую последовательность пустых строк.Для этого используется регулярное выражение ‘\s*\n\s*\n\s*’
- **Плюсы**: разбивает только по символам пробела или табуляции.
- **Минусы**: .

```python
tok = tokenize.BlanklineTokenizer()
print(tok.tokenize(text))
```

### 2 - LegalitySyllableTokenizer
Разбивает слова по слогам, основываясь на принципе правильности и максимизации наборов.
- **Плюсы**: Реализация указанного принципа в библиотеке NLTK по умолчанию предполагает набор английских гласных, но атрибут vowel позволяет установить набор гласных букв любого другого алфавита для использования
- **Минусы**: 
    1. не работает одинаково для разных языков. 
    2. Этот токенайзер работает с отдельными словами, поэтому исходный текст нужно предварительно обработать. 
    3. также для корректной работы токенайзера потребуется словарь слов русского языка и список гласных букв, передаваемый через параметр vowel.

```python
with open('BowenJamesASteetCatNamedBob.txt', 'r', encoding='utf-8') as f:
    text = f.readlines()

clear_words = [line.split('\t')[0].lower() for line in text if len(line.split('\t')) > 1]
print(len(clear_words))
```

### 3 - LineTokenizer
Разбивает строку на токены, использует символ новой строки(похоже на s.split(‘\n’)).
- **Плюсы**: 
    1. убирает пустые строки.
    2. при необходимости может отбрасывать пустые строки.
- **Минусы**:

```python
tok = tokenize.LineTokenizer()
tok.tokenize(text)
```

### 4 - MWETokenizer -- (Multi-Word Expression Tokenizer)
MWETokenizer берет строку, которая уже была разделена на токены, и повторно токенизирует ее, объединяя выражения из нескольких слов в отдельные токены, используя словарь MWE. 
- **Плюсы**: удобно, когда в тексте встречаются устойчивые выражения или именованные сущности, состоящие из нескольких слов.
- **Минусы**:

```python
txt = 'There’s a famous quote I read somewhere. It says we are all given second chances every day of our lives. They are there for the taking, it’s just that we don’t usually take them.'
tok = tokenize.WordPunctTokenizer()
text_words = tok.tokenize(txt)
print(len(text_words), text_words)
tok = tokenize.MWETokenizer(separator=' ')
tok.add_mwe(('famous', 'somewhere'))
print(len(tok.tokenize(text_words)), tok.tokenize(text_words))
```


### 5 - NLTKWordTokenizer
Улучшенный вариант TreebankWordTokenizer, включающий несколько адаптированных списков сокращений.
- **Плюсы**: для простоты нужно загрузить предварительно обученную модель из nltk  для токенизации текста.
- **Минусы**: является «деструктивным» токенизатором, регулярные выражения приводят исходный текст в состояние, не поддающееся восстановлению

```python
txt = tokenize.LineTokenizer()
txt = txt.tokenize(text)[5]
tok = tokenize.NLTKWordTokenizer()
tok.tokenize(txt)
```

### 6 - PunktSentenceTokenizer
Рзделяет текст на список предложений, используя алгоритм обучения без учителя (Unsupervised learning) для построения модели слов-сокращений, словосочетаний и слов, с которых начинаются предложения.
- **Плюсы**: для простоты нужно загрузить предварительно обученную модель из nltk  для токенизации текста.
- **Минусы**: прежде чем использовать, его необходимо обучить на большом наборе текстов, написанных на целевом языке.

```python
txt = tokenize.LineTokenizer()
txt = txt.tokenize(text)[5]
tok = nltk.data.load('tokenizers/punkt/russian.pickle')
print(len(tok.tokenize(txt)))
print(tok.tokenize(txt))
```

### 7 - RegexpTokenizer
Разбивает исходный текст на подстроки, используя регулярное выражение, передаваемое ему в качестве параметра (например: выбор из текста только слова, написанные с заглавной буквы). 
1. pattern – шаблон регулярного выражения, используемый для создания токенизатора;
2. gaps – параметр, определяющий, будет ли токенизатор искать сами токены (False) или разделители между ними (True). В первом случае результат работы токенизатора для указанного примера будет аналогичен выполнению команды re.findall(pattern, text) с использованием библиотеки re;
3. discard_empty – необходимо установить в значение True, если необходимо отбросить пустые токены, сгенерированные токенизатором (возможно только в случае gaps == True);
4. flags – флаги регулярного выражения, используемы при компиляции шаблона. По умолчанию установлены 3 флага: re.UNICODE | re.MULTILINE | re.DOTALL.

- **Плюсы**: удобен при выборке.
- **Минусы**: не всегда корректен в поиске.


```python
pattern = '[A-Z]\w+'
tok = tokenize.RegexpTokenizer(pattern)
print(tok.tokenize(text))
```

### 9 - SExprTokenizer
Используется для поиска в исходном тексте выражений в скобках.
- **Плюсы**: делит строку на последовательность, вызывает исключение ValueError - если используется для токенизации выражения с непарными круглыми скобками. Удобно использовать для проверки правильной расстановки скобок в математических выражениях, а также при парсинге и обработке тестов, содержащих большое количество формул, выделяемых круглыми скобками или иными символами.
- **Минусы**: включет любые вложенные выражения в скобках. Любая непарная закрывающая скобка будет выделена в качестве самостоятельного токена(но непарная открывающая круглая скобка будет выделена в качестве токена вместе со всеми символами, оставшимися до конца строки).отдельная обработка для исключения скобок, встречающихся внутри уже выделенных токенов, или следующих за символом обратной косой черты, не производится

В моем тексте нет, выражений в скобках(

```python
text_example = '(a + b / (2 + y)) * c - x) + 25 / (87'
tok = tokenize.SExprTokenizer(strict=False)
print(len(tok.tokenize(text_example)))
print(tok.tokenize(text_example))
```

### 14 - TextTilingTokenizer
Автоматически разделяет полноразмерных текстовых документов на последовательные блоки, состоящие из нескольких абзацев, которые представляют собой отрывки или подтемы. 
- **Плюсы**: подтемы обнаруживаются на основе количественного лексического анализа закономерностей совместной встречаемости. После анализа результатов можно выбрать наиболее подходящее для решаемой задачи сочетание параметров.
- **Минусы**: не всегда соответствует ожидаемому результату.

```python
stopwords = nltk.corpus.stopwords.words('english')
from itertools import product
w = [1, 2, 3, 4]
k = [1, 2, 3]
for wi, ki in product(w, k):
    tok = tokenize.TextTilingTokenizer(w=wi, k=ki, stopwords=stopwords)
    try:
        text_words = tok.tokenize(text)
        if len(text_words) > 1:
            print(wi, ki, len(text_words))
    except:
        pass
```

### 15 - ToktokTokenizer
На вход которого подается по одному предложению в строке.
- **Плюсы**: Входные данные должны быть представлены в кодировке UTF-8. Замена неразрывных пробелов обычными, замена знака многоточия единичной точкой, расстановка пробелов после открывающих и перед закрывающими знаками препинания.
- **Минусы**: нужно проверять и дополнительно очищать текст после применения , от знаков препинания и других символов – бывает, что они «приклеиваются» к словам.

```python
txt = tokenize.LineTokenizer()
txt = txt.tokenize(text)[0]
tok = tokenize.ToktokTokenizer()
tok.tokenize(txt)
```


### 17,18 - TreebankWordTokenizer
Аннотирует синтаксическую или семантическую структуру предложения. Использует регулярные выражения для разделения текста на слова.
Tokenizer можно инвертировать с помощью TreebankWordDetokenizer, который использует обратные операции регулярных выражений, соответствующих регулярным выражениям токенизатора Treebank.
- **Плюсы**:большинство знаков препинания, разделяются на отдельные токены, отделяются точки, которые проставлены в конце строк.
- **Минусы**: разделяются стандартные сокращения, например, «don’t» -> »do n’t» или «they’ll» -> «they ‘ll» для английского языка;

```python
txt = tokenize.LineTokenizer()
txt = txt.tokenize(text)[5]
tok = tokenize.TreebankWordTokenizer()
print(list(tok.span_tokenize(txt)))
```
